```python
import branca, folium, pandas as pd, os
csvloc = 'csv'
dirr = os.listdir(csvloc)
dirr
```

```python
df = pd.read_csv('loc.csv')
m = folium.Map(location = [14.6760,121.0437], zoom_start=12)
```

```python
for k,v in df.iterrows():
    data = pd.read_csv(os.path.join(csvloc,v.tag+'.csv'))
    html = """
            <h5> {} </h5>
           """.format(v['name'])
    if 't' in list(data.triggered):
        folium.Marker(
            location = [v.lat,v.lon],
            popup = html,
            icon = folium.Icon(color='red')
        ).add_to(m)
    else:
        folium.Marker(
            location = [v.lat,v.lon],
            popup = html,
            icon = folium.Icon(color='green')
        ).add_to(m)
```

```python
with open('templates/plpotly.html','rb') as h:
    html = """
            <h5> {} </h5>
           """
    iframe = branca.element.IFrame(html=h.read(),width=530,height=400)
    popup = folium.Popup(iframe,max_width=2650)
    folium.Marker(
        location = [14.6760,121.0437],
        popup = popup,
        icon=folium.Icon(color='red')
    ).add_to(m)
```
