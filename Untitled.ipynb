```python
import branca, folium, pandas as pd
```

```python
df = pd.read_csv('loc.csv')
m = folium.Map(location = [14.6760,121.0437], zoom_start=12)
```

```python
for k,v in df.iterrows():
    folium.Marker(
        location = [v.lat,v.lon],
        popup = v['name'],
        icon = folium.Icon(color='green')
    ).add_to(m)
```
